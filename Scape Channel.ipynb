```python
import requests
from bs4 import BeautifulSoup
import csv
import json
import re
```

```python
def scrape_google_results(search_query, num_results,cookies,headers):
    start=0
    links=[]
    channel=[]
    for i in range(0,10):
        url=f"https://www.google.com/search?q={search_query}&num={num_results}&sxsrf=APwXEdenbbYEP7byNfEQ5B3EkMncTG_ZYA:1684348406647&ei=9h1lZLaPJ8iN4-EPj9qI2Aw&start={start}&sa=N&filter=0&ved=2ahUKEwi24tLj_vz-AhXIxjgGHQ8tAss4rAIQ8tMDegQIBRAE&biw=978&bih=714&dpr=1.25"
        response = requests.get(
            url=url,
            cookies=cookies,
            headers=headers,
        )
        print(response)
        soup = BeautifulSoup(response.text, 'html.parser')
        results = soup.find_all('div', {'class': 'DhN8Cf'})
        # channel_result=soup.find_all('div', {'class': 'P7xzyf'})
        # text = channel_result.find_all(text=True, recursive=False)
        # print(text)
        # text = ' '.join(text).strip()
        # print(text)
        div = soup.find_all('div', class_='P7xzyf')
        # Extract the desired text
        for ele in div:
            p=str(ele)
            cha=p.split(" Â· ")[1]
            s= BeautifulSoup(cha, 'html.parser')
            f = s.find('span').text
            f=re.sub(r'[^A-Za-z]', '', f)
            link="https://www.youtube.com/@"+f
            channel.append(link)
            # text = div_element.find_all(text=True, recursive=False)
            # print(text)
            # text = ' '.join(text).strip()
            # print(text)
        # print(channel_result)
        # print(results)

        for result in results:
            link = result.find('a')['href']
            if "youtube.com" in link:
                links.append(link)
        start+=100

    return links,channel
```

```python
def save_links_to_csv(links):
    with open('youtube_link.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['YouTube Channel Links'])
        writer.writerows([[link] for link in links])

def save_channel_to_csv(links):
    with open('youtube_channels.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['YouTube Channel Links'])
        writer.writerows([[link] for link in links])

def save_links_to_json(links):
    data = {'YouTube Channel Links': links}
    with open('youtube_link.json', 'w', encoding='utf-8') as file:
        json.dump(data, file, indent=4)
```

```python
if __name__ == "__main__":
    search_query = "site:youtube.com openinapp.co"
    num_results = 10000

    # Scrape Google search results
    links,channels = scrape_google_results(search_query, num_results,cookies,headers)

    # Save links to CSV file
    save_links_to_csv(links)

    save_channel_to_csv(channels)
    # Save links to JSON file
    save_links_to_json(links)

    print(f"Scraped {len(links)} YouTube links and saved to 'youtube_link.csv' and 'youtube_link.json' files.")
    print(f"Scraped {len(channels)} YouTube channel links and saved to 'youtube_channels.csv' file.")
```
